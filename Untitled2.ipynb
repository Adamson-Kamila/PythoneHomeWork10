```python
lst = ['robot'] * 10 
lst += ['human'] * 10 
random.shuffle(lst) 
data = pd.DataFrame({'whoAmI':lst}) 
data.head(n=20) 
```

```python
data.loc[data['whoAmI'] =='human', ['human', 'robot']]=[1, 0]
data.loc[data['whoAmI'] =='robot', ['human', 'robot']]=[0, 1]
data.head(n=20) 
```

```python
sb.heatmap(data[['human', 'robot']])
```
